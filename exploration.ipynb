# Drug Consumption in America Visualization

This notebook explores and visualizes drug consumption data in America.

## Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

## Load and Explore the Dataset

```python
# Load the dataset from CDC
url = "https://data.cdc.gov/api/views/xkb8-kh2a/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# Display first few rows
df.head()

# Get info
df.info()

# Describe numerical columns
df.describe()
```

## Data Cleaning and Preprocessing

```python
# Filter for drug overdose deaths
df = df[df['Indicator'] == 'Number of Drug Overdose Deaths']

# Convert Data Value to numeric
df['Data Value'] = pd.to_numeric(df['Data Value'], errors='coerce')

# Drop rows with missing values in key columns
df.dropna(subset=['Data Value', 'Year', 'State'], inplace=True)

# Display cleaned data
df.head()
```

## Exploratory Data Analysis

```python
# Group by year and sum deaths
yearly_deaths = df.groupby('Year')['Data Value'].sum().reset_index()
print("Yearly Deaths:")
print(yearly_deaths)

# Group by state
state_deaths = df.groupby('State')['Data Value'].sum().reset_index().sort_values('Data Value', ascending=False)
print("Top States by Deaths:")
print(state_deaths.head(10))

# Correlation if applicable
# For time series, perhaps rolling averages
```

## Create Static Visualizations

```python
# Line plot for yearly deaths
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_deaths, x='Year', y='Data Value')
plt.title('Drug Overdose Deaths in America Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.show()

# Bar plot for top states
plt.figure(figsize=(12, 8))
sns.barplot(data=state_deaths.head(10), x='Data Value', y='State')
plt.title('Top 10 States by Drug Overdose Deaths')
plt.xlabel('Number of Deaths')
plt.ylabel('State')
plt.show()
```

## Create Interactive Visualizations

```python
# Interactive line plot
fig = px.line(yearly_deaths, x='Year', y='Data Value', title='Drug Overdose Deaths in America Over Time')
fig.show()

# For choropleth, we need state codes
# Assuming df has 'State' as full names, map to abbreviations
state_abbrev = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
    'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
    'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
    'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}
state_deaths['State Code'] = state_deaths['State'].map(state_abbrev)

fig2 = px.choropleth(state_deaths, locations='State Code', locationmode='USA-states', color='Data Value',
                     scope='usa', title='Drug Overdose Deaths by State')
fig2.show()
```
